# Inserting Data Into Elastic
## Below two steps in Kibana
- Delete Old Index

`DELETE sentence-index`

- Create new index

`PUT sentence-index
{
  "mappings": {
    "properties": {
      "sentence-vector": {
        "type": "dense_vector",
        "dims": 384,
        "index": false
      },
      "sentence": {
        "type": "text"
      }
    }
  }
}`

```python
for id, row in df.iterrows():
    doc = { "sentence-vector": [], "sentence": "" }
    print(id)
    doc['sentence-vector'] = list(row['embedding'])
    doc['sentence'] = row['review']
    resp = es.index(index="sentence-index", id=id, document=doc)
    
```
